```python
import pickle
import matplotlib.pyplot as plt
from pathlib import Path
import os

current_file_dir = Path(os.getcwd())
result_dir = current_file_dir.parents[1] / "results"
run_dir = result_dir / "white-shiny-hawk-20250224-0606"
adv_file = "model_adv.pkl"
base_file = "model_base.pkl"

def modify_rs_plot(file_pth):
    with open(file_pth,'rb') as file:
        rs_plot = pickle.load(file)
        plt.xlabel("Radius")
        plt.ylabel("Certified Accuracy")
        lines = plt.gca().get_lines()

        colors = ["#007FFF", "#FF6F00", "#D72638"]
        # colors = ["#E63946", "#457B9D", "#2A9D8F"]  # Pick from above
        for line, color in zip(lines, colors):
            line.set_color(color)
        labels = [line.get_label() for line in lines]  # Get original labels
        plt.legend(lines, labels)  # Reassign legend with updated colors
        plt.show()

modify_rs_plot(run_dir / adv_file)
```
